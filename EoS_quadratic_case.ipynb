```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return 0.5 * x**2

def grad(x):
    return x

def simulate_emahb(eta, beta1 = 0.9, x0 = 5.0, steps = 100):
    m = 0
    x = x0
    x_history = [x0]
    
    for n in range(1, steps):
        m = beta1 * m + (1 - beta1) * grad(x)
        x = x - eta * m
        x_history.append(x)

    return x_history
```

```python
etas = [1, 37, 37.9, 38.1, 39]
all_loss_paths = []
for eta in etas:
    loss = f(np.array(simulate_emahb(eta)))
    all_loss_paths.append(loss)
    plt.plot(loss, label=f' = {eta}')

plt.xlabel('Step')
plt.xlim(0, 110)
plt.ylabel('Loss')
plt.yscale('log')
plt.legend()
plt.show()
```
